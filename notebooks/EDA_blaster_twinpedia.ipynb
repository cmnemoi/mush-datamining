```python
from tqdm import tqdm

import numpy as np
import pandas as pd
```

```python
blaster_data = pd.read_csv('../data/blaster_twinpedia.csv', parse_dates=["Horodateur"])
```

```python
blaster_data
```

```python
fails = ["poney imaginaire", "tour de magie", "est enrayé", "a raté une occasion", "se retourne subitement"]
```

```python
blaster_data["Touché"] = blaster_data["Message"].apply(lambda x: not any(fail in x for fail in fails))
```

```python
blaster_data
```

```python
successful_shots = blaster_data[blaster_data["Touché"] == True]
```

```python
# merge "Plaie 1", "Plaie 2", "Plaie 3" into one column
successful_shots["Plaies"] = successful_shots["Plaie 1"].fillna("") + "," + successful_shots["Plaie 2"].fillna("") + "," + successful_shots["Plaie 3"].fillna("")
successful_shots["Plaies"].replace("", pd.NA, inplace=True)
injuries = successful_shots["Plaies"].str.split(",", expand=True).stack().reset_index(level=1, drop=True).rename("Plaies")
injuries = injuries[injuries != ""]
```
