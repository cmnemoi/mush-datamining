```python
from tqdm import tqdm

import numpy as np
import pandas as pd
```

```python
couteau_data = pd.read_csv('../data/couteau_twinpedia.csv', parse_dates=["Horodateur"])
```

```python
couteau_data
```

```python
fails = ["perds son couteau", "se mordre la langue", "tapis imaginaire", "claque un muscle dorsal", "Trop vif", "Zip", "dégaine sa lame"]
```

```python
couteau_data["Touché"] = couteau_data["Message"].apply(lambda x: not any(fail in x for fail in fails))
```

```python
couteau_data
```

```python
successful_blows = couteau_data[couteau_data["Touché"] == True]
```

```python
# merge "Plaie 1", "Plaie 2", "Plaie 3" into one column
successful_blows["Plaies"] = successful_blows["Plaie 1"].fillna("") + "," + successful_blows["Plaie 2"].fillna("") + "," + successful_blows["Plaie 3"].fillna("")
successful_blows["Plaies"].replace("", pd.NA, inplace=True)
injuries = successful_blows["Plaies"].str.split(",", expand=True).stack().reset_index(level=1, drop=True).rename("Plaies")
injuries = injuries[injuries != ""]
```

```python
injuries.value_counts(normalize=True) * 100
```

```python
len(injuries) / len(successful_blows) * 100
```

```python
failed_blows = couteau_data[couteau_data["Touché"] == False]
failed_blows["Plaies"] = failed_blows["Plaie 1"].fillna("") + "," + failed_blows["Plaie 2"].fillna("") + "," + failed_blows["Plaie 3"].fillna("")
failed_blows["Plaies"].replace("", pd.NA, inplace=True)
agressor_injuries = failed_blows["Plaies"].str.split(",", expand=True).stack().reset_index(level=1, drop=True).rename("Plaies")
agressor_injuries = agressor_injuries[agressor_injuries != ""]
```

```python
agressor_injuries.value_counts(normalize=True) * 100
```

```python
len(agressor_injuries) / len(failed_blows) * 100
```
